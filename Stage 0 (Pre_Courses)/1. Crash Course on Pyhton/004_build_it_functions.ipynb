We already explored built-in functions like `print()`, `type()` and `str()`.
**Built-in** functions are functions that exist in Python and can be called directly.
Let's see this further and take a look at `sorted()`, `max()`, `min()`, and see how to pass value from one to another function.

### print()

- Outputs specified object to the screen.
- One of the most commonly used functions in Python, not just for functionality but for testing also.
- `print()` takes any number of arguments, separated by comma, and prints all of them.
- you can run code that prints string, float, int, etc. together.

```python
month = 'September'
print("Investigate failed login attempts in month of", month, "if more then", 100)
```

### type()

- Returns data type of its argument.
- It helps us to keep track of data types and avoid errors throughout our code.
- `type()` only takes one argument and returns its type.

```python
type("what type is this")
```

### Passing one function into another

>consider next code block

```python
print(type('some string'))
```

It displays str(), which means that the argument passed to the type() function is a string. This happens because the type() function is processed first and its output is passed as an argument to the print() function.

### str()

- Can be used to convert `any data type into string`.
- It takes single argument, and returns string representation of that argument.

```python
number = 12
string_representation = str(number)
print(string_representation)
```

### sorted()

- Sorts components of a list.
- `sorted()` function also works on any `iterable`, like a string, and returns items in a list.
- By default it sorts them in `ascending` order.
    - If given iterable that contains numbers it sorts them from smallest to largest.
    - Iterables containing string will be sorted alphabetically
- `It does not change iterable it is sorting`
- It cannot take 

```python
time_list = [12, 2, 32, 19, 57, 22, 14]
print(sorted(time_list))
print(time_list)
```

### min() and max()

- The `max()` function returns the largest numeric input passed into it.
- The `min()` function returns the smallest numeric input passed into it.

- The `max()` and `min()` functions accept arguments of **either** multiple numeric values **or** of an iterable like a list, and they return the largest or smallest value respectively.


```python
time_list = [12, 2, 32, 19, 57, 22, 14]
print(min(time_list))
print(max(time_list))
```
